```python
dataset.columns = ['States', 'Date', 'Frequency', 'Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate', 'Region']
```

```python

dataset['Frequency'] = dataset['Frequency'].astype('category')
dataset['Region'] = dataset['Region'].astype('category')
```

```python

dataset['Date'] = pd.to_datetime(dataset['Date'], dayfirst=True)
dataset['Month'] = dataset['Date'].dt.month
dataset['Month_int'] = dataset['Month'].apply(lambda x: int(x) if pd.notna(x) else x)
```

```python

dataset.drop(columns='Month', inplace=True)
```

```python
dataset_ld = dataset.copy()
```

```python
dataset=dataset.dropna()
```

```python

mask = (dataset['Date'] >= '2019-05-31') & (dataset['Date'] <= '2020-03-31')
dataset = dataset.loc[mask]
```

```python

mask = (dataset_ld['Date'] >= '2020-04-30')
dataset_ld = dataset_ld.loc[mask]
```

```python
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
sns.barplot(x='Region', y='Estimated Unemployment Rate', data=dataset)
plt.xlabel("Region", fontsize=10)
plt.ylabel('Estimated Unemployment Rate', fontsize=10)
plt.title('Estimated Unemployment Rate Before Lockdown', fontsize=12)
plt.subplot(1, 2, 2)
sns.barplot(x='Region', y='Estimated Unemployment Rate', data=dataset_ld)
plt.xlabel("Region", fontsize=10)
plt.ylabel('Estimated Unemployment Rate', fontsize=10)
plt.title('Estimated Unemployment Rate During Lockdown', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(25, 8))
plt.subplot(1, 2, 1)
plt.title("State Wise Estimated Unemployment Rate Before Lockdown", fontsize=16)
sns.swarmplot(y="States", x='Estimated Unemployment Rate', data=dataset, size=8)
plt.xlabel("Estimated Unemployment Rate", fontsize=16)
plt.ylabel("States", fontsize=16)
plt.subplot(1, 2, 2)
plt.title("State Wise Estimated Unemployment Rate During Lockdown", fontsize=16)
sns.swarmplot(y="States", x='Estimated Unemployment Rate', data=dataset_ld, size=8)
plt.xlabel("Estimated Unemployment Rate", fontsize=16)
plt.ylabel("States", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(25, 8))
plt.subplot(1, 2, 1)
plt.title("Estimated Unemployment Rate Before Lockdown", fontsize=16)
sns.barplot(data=dataset.sort_values(by='Estimated Unemployment Rate', ascending=False), x="States", y="Estimated Unemployment Rate")
plt.xticks(rotation=90)
plt.xlabel('States', fontsize=16)
plt.ylabel('Estimated Unemployment Rate', fontsize=16)
plt.subplot(1, 2, 2)
plt.title("Estimated Unemployment Rate During Lockdown", fontsize=16)
sns.barplot(data=dataset_ld.sort_values(by='Estimated Unemployment Rate', ascending=False), x="States", y="Estimated Unemployment Rate")
plt.xticks(rotation=90)
plt.xlabel('States', fontsize=16)
plt.ylabel('Estimated Unemployment Rate', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(25, 8))
plt.subplot(1, 2, 1)
plt.title("Estimated Employed Workforce Before Lockdown", fontsize=16)
sns.barplot(data=dataset.sort_values(by='Estimated Employed', ascending=False), x="States", y="Estimated Employed")
plt.xticks(rotation=90)
plt.xlabel('States', fontsize=16)
plt.ylabel('Estimated Employed', fontsize=16)
plt.subplot(1, 2, 2)
plt.title("Estimated Employed Workforce During Lockdown", fontsize=16)
sns.barplot(data=dataset_ld.sort_values(by='Estimated Employed', ascending=False), x="States", y="Estimated Employed")
plt.xticks(rotation=90)
plt.xlabel('States', fontsize=16)
plt.ylabel('Estimated Employed', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(25, 8))
plt.subplot(1, 2, 1)
plt.title("Estimated Labour Participation Rate Before Lockdown", fontsize=16)
sns.barplot(data=dataset.sort_values(by='Estimated Labour Participation Rate', ascending=False), x="States", y="Estimated Labour Participation Rate")
plt.xticks(rotation=90)
plt.xlabel('States', fontsize=16)
plt.ylabel('Estimated Labour Participation Rate', fontsize=16)
plt.subplot(1, 2, 2)
plt.title("Estimated Labour Participation Rate During Lockdown", fontsize=16)
sns.barplot(data=dataset_ld.sort_values(by='Estimated Labour Participation Rate', ascending=False), x="States", y="Estimated Labour Participation Rate")
plt.xticks(rotation=90)
plt.xlabel('States', fontsize=16)
plt.ylabel('Estimated Labour Participation Rate', fontsize=16)
plt.show()
```

```python
df1_stats = dataset[['Estimated Unemployment Rate','Estimated Employed', 'Estimated Labour Participation Rate']]

df2_stats = dataset_ld[['Estimated Unemployment Rate','Estimated Employed', 'Estimated Labour Participation Rate']]
     
```
